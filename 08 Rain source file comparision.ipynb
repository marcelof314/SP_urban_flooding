```python
rain_source_marcelo = '/Users/marcelofernandes/My Drive/Arquivos/Rain_source_2023_1031.csv'
```

```python
df_rain_source_marcelo = pd.read_csv(rain_source_marcelo)
```

```python
df_rain_source_marcelo.info()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o DataFrame (substitua pelo seu DataFrame real)
df = df_rain_source_marcelo  # Seu DataFrame carregado

# Ajustar a data inicial (01 de janeiro de 2023)
start_date = pd.Timestamp('2023-01-01')
df['datetime'] = start_date + pd.to_timedelta(df['0'], unit='s')  # Converter '0' para datetime
df.set_index('datetime', inplace=True)  # Configurar datetime como índice


# Criar colunas de hora e dia
df['hour'] = df.index.hour
df['day'] = df.index.date

# Escolher uma coluna de interesse para análise (exemplo: coluna '1')
col_of_interest = '1'  # Substitua pela coluna desejada
heatmap_data = df.groupby(['day', 'hour'])[col_of_interest].mean().unstack()

# Criar o heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='viridis', cbar_kws={'label': 'Valor'})
plt.title('Heatmap Diário por Hora')
plt.xlabel('Hora do Dia')
plt.ylabel('Dia')
plt.show()
```
